```python
players_defending=fifa_df[['defending','short_name','club','nationality']]
```

###  Best defenders in football currently

```python
players_defending.sort_values(by=['defending'],ascending=False).head(10)
```

```python
a=players_defending.defending.mean()+ 2*(players_defending.defending.std())
```

```python
a
```

```python
b=players_defending[players_defending['defending'] > a]
```

```python
b=b.sort_values(by=['defending'],ascending=False).head(50)
b
```

```python
country_defending=b
country_defending.drop(['short_name','club'],axis=1,inplace=True)
```

###  Country-wise defending ability of teams

```python
country_defending.groupby(['nationality']).count().sort_values(by=['defending'],ascending=False).plot.bar()
```

```python
players_finishing=fifa_df[['shooting','short_name','club','nationality']]
```

###  Players with Best finishing ability

```python
players_finishing.sort_values(by=['shooting'],ascending=False).head(10)
```

```python
c=fifa_df['shooting'].mean()+ 2*(fifa_df['shooting'].std())
```

```python
players_finishing=fifa_df[['shooting','short_name','club','nationality']]
players_finishing=players_finishing[players_finishing['shooting']>c]
players_finishing.head(50)
```

### Country-wise shooting ability of teams

### Best Central midfielders

###  European teams dominate with respect to passing ability

###  Visualization on  the dataset to get insights

```python
%matplotlib inline 
plt.pie(df.OrderType.value_counts(),labels=df.OrderType.value_counts().keys(),autopct='%0.1f%%')
plt.title('Market-wise distribution')
plt.show()
```

### Region-wise sales analysis based on different parameters

```python
regionwise_sales=df[['OrderType','Sales Region','Order Total']]
regionwise_sales.groupby(['Sales Region','OrderType']).agg(['sum'])
```

```python
regionwise_sales.groupby(['Sales Region','OrderType']).sum().sort_values(by=['Order Total'],ascending=False).plot.bar()
```

### State-wise sales analysis on various parameters

```python
statewise_sales=df[['CustState','Order Total']]
statewise_sales.groupby(['CustState']).sum().sort_values(by=['Order Total'],ascending=False).head(15).plot.bar(color='green')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(df.ProdCategory.value_counts(),labels=df.ProdCategory.value_counts().keys(),autopct='%0.1f%%')
plt.show()
```
